```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

# Load the data from the uploaded CSV file
file_path = '/content/Penyebab Kematian di Indonesia yang Dilaporkan - Clean.csv'
data = pd.read_csv(file_path)

# Data preparation: group by 'Cause' and 'Year' to get total deaths per cause each year
cause_trend = data.groupby(['Cause', 'Year'])['Total Deaths'].sum().reset_index()

# We will select some major causes of death for visualization purposes
major_causes = cause_trend.groupby('Cause')['Total Deaths'].sum().nlargest(5).index
filtered_data = cause_trend[cause_trend['Cause'].isin(major_causes)]

# Set the plot size
plt.figure(figsize=(10, 6))

# Plot the trend of deaths by major causes
sns.lineplot(data=filtered_data, x='Year', y='Total Deaths', hue='Cause', marker='o')

# Adding titles and labels
plt.title('Kematian di Indonesia Berdasarkan Beberapa Penyebab', fontsize=14)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Jumlah Kematian', fontsize=12)
plt.legend(title='Penyebab', bbox_to_anchor=(1.05, 1), loc='upper left')

# Ensure the x-axis only shows whole years (no decimals)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(integer=True))

# Show the plot
plt.tight_layout()
plt.show()
```
